```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import matplotlib
```

```python
# DOS bulk slab: PBE-ONCV

efermi_1 = 5.8696
dostxtf_1 = '/home/agarcia/Arturo_MATS/Hexaborides_project/CaB6/slab_supercells/1x2_slab/4_vacuum/PDOS_Mulliken/PBE-ONCV/bulk-slab/DOS/bulk_slab_dos.txt'

data_1 = np.loadtxt(dostxtf_1)
energy_1, dos_1, dos_int = data_1.transpose()
x_1 = energy_1 - efermi_1

plt.figure(figsize=(9, 5))
plt.plot(x_1, dos_1, linewidth=0.75, color='k', label='DOS (total PDOS) - bulk: ONCV')
plt.yticks([])
plt.xlabel('Energy (E - E$_F$) (eV)')
plt.ylabel('DOS')
plt.axvline(x=0, linewidth=0.5, color='k', linestyle=(0, (8, 10)))
plt.xlim(-13 - efermi_1, 15 - efermi_1)
plt.ylim(0, 150)
plt.fill_between(x_1, 0, dos_1, where=(x_1 < 0), facecolor='k', alpha=0.25)
plt.text(0 + 0.2, 50.7, 'Fermi energy', rotation=90)
plt.legend(frameon=False, loc='upper left')
#plt.savefig('bulk_DOS_ONCV.png', dpi=400)
plt.show()
```
