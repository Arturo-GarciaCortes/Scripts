```python
import matplotlib.pyplot as plt
from matplotlib import rcParamsDefault
import numpy as np

# load data
def data_loader(fname):
    import numpy as np

    data = np.loadtxt(fname)
    energy = data[:, 0]
    pdos = data[:, 1]  # ldos col, total contribution for a given orbital

    return energy, pdos
```

```python
# PDOS bulk slab: PBE-PSlibray

efermi_1 = 5.8834

# PDOS of atom Ca #1 (surface atom)
energy_1, Ca_pdos_1s = data_loader('/home/agarcia/Arturo_MATS/Hexaborides_project/CaB6/slab_supercells/1x2_slab/4_vacuum/PDOS_Mulliken/PBE_PSlibray/bulk_slab/PDOS/bulk_slab_pdos.dat.pdos_atm#1(Ca)_wfc#1(s)')
_, Ca_pdos_2s = data_loader('/home/agarcia/Arturo_MATS/Hexaborides_project/CaB6/slab_supercells/1x2_slab/4_vacuum/PDOS_Mulliken/PBE_PSlibray/bulk_slab/PDOS/bulk_slab_pdos.dat.pdos_atm#1(Ca)_wfc#2(s)')
_, Ca_pdos_3p = data_loader('/home/agarcia/Arturo_MATS/Hexaborides_project/CaB6/slab_supercells/1x2_slab/4_vacuum/PDOS_Mulliken/PBE_PSlibray/bulk_slab/PDOS/bulk_slab_pdos.dat.pdos_atm#1(Ca)_wfc#3(p)')
_, Ca_pdos_4p = data_loader('/home/agarcia/Arturo_MATS/Hexaborides_project/CaB6/slab_supercells/1x2_slab/4_vacuum/PDOS_Mulliken/PBE_PSlibray/bulk_slab/PDOS/bulk_slab_pdos.dat.pdos_atm#1(Ca)_wfc#4(p)')

x_1 = energy_1 - efermi_1

# make plots
plt.figure(figsize = (8, 4))
plt.plot(x_1, Ca_pdos_1s, linewidth=0.75, color='y', label='Ca 1s-orbital')
plt.plot(x_1, Ca_pdos_2s, linewidth=0.75, color='blue', label='Ca 2s-orbital')
plt.plot(x_1, Ca_pdos_3p, linewidth=0.75, color='r', label='Ca 3p-orbital')
plt.plot(x_1, Ca_pdos_4p, linewidth=0.75, color='green', label='Ca 4p-orbital')
plt.yticks([])
plt.xlabel('Energy (E - E$_F$) (eV)')
plt.ylabel('DOS')
plt.axvline(x=0, linewidth=0.5, color='k', linestyle=(0, (8, 10)))
plt.xlim(2 - efermi_1, 12 - efermi_1)
plt.ylim(0, 2)
plt.fill_between(x_1, 0, Ca_pdos_1s, where=(x_1 < 0), facecolor='y', alpha=0.25)
plt.fill_between(x_1, 0, Ca_pdos_2s, where=(x_1 < 0), facecolor='blue', alpha=0.25)
plt.fill_between(x_1, 0, Ca_pdos_3p, where=(x_1 < 0), facecolor='r', alpha=0.25)
plt.fill_between(x_1, 0, Ca_pdos_4p, where=(x_1 < 0), facecolor='green', alpha=0.25)
plt.text(0.1, 1.2, 'Fermi energy', rotation=90)
plt.legend(frameon=False)
#plt.savefig('PDOS_bulk_Ca#1.png', dpi=400)
plt.show()
```

```python
# PDOS total bulk slab (DOS): PBE-PSlibrary

_, pdos_tot = data_loader('/home/agarcia/Arturo_MATS/Hexaborides_project/CaB6/slab_supercells/1x2_slab/4_vacuum/PDOS_Mulliken/PBE_PSlibray/bulk_slab/PDOS/bulk_slab_pdos.dat.pdos_tot')
plt.plot(x_1, pdos_tot, linewidth=0.75, color='k', label='total')
plt.fill_between(x_1, 0, pdos_tot, where=(x_1 < 0), facecolor='k', alpha=0.25)
plt.yticks([])
plt.xlabel('Energy (E - E$_F$) (eV)')
plt.ylabel('DOS')
plt.axvline(x=0, linewidth=0.5, color='k', linestyle=(0, (8, 10)))
plt.xlim(-10 - efermi_1, 15 - efermi_1)
plt.ylim(0.0, 150)
plt.text(0.4, 40, 'Fermi energy', rotation=90)
plt.legend(frameon=False)
#plt.savefig('total_PDOS_bulk.png', dpi=400)
plt.show()
```
